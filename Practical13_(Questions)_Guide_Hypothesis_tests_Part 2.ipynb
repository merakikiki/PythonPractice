## Q1

A company claims that their new training program significantly improves employee productivity. To test this claim, a random sample of 30 employees is selected. Half of the employees are randomly assigned to undergo the new training program, while the other half does not receive any training. After the training period, the productivity scores (in units produced per hour) for both groups are recorded as follows:

productivity scores (new training): 12,14,10,11,13,15,9,12,13,16,14,12,11,12,10,13,12,15,14,13

productivity scores (no training):10,9,8,11,9,10,12,8,11,9,10,12,10,11,8,9,11,10,12,9


Perform a two-sample t-test to determine if there is a significant difference in the mean productivity scores between the two groups. Use a 5% level of significance.

#### Solution:

#### Define your hypothesis    

H0: There is no significant difference in the mean productivity scores between employees who underwent the new training program and those who did not.
H1: There is a significant difference in the mean productivity scores between employees who underwent the new training program and those who did not.

#### Put our group in lists

```python
new_training = [12, 14, 10, 11, 13, 15, 9, 12, 13, 16, 14, 12, 11, 12, 10, 13, 12, 15, 14, 13]
no_training = [10, 9, 8, 11, 9, 10, 12, 8, 11, 9, 10, 12, 10, 11, 8, 9, 11, 10, 12, 9]
```

#### Check the sample distrbuition using kde plot

```python
#Let us check their distrbuition
sns.kdeplot(new_training, color="yellow", fill=True)
sns.kdeplot(no_training, color="purple", fill=True)

plt.show()
```

#### Use scipy.stats to perform two sample t-test

```python
# Two Sample t - test
from scipy.stats import ttest_ind

t_statistic, p_value = ttest_ind(new_training, no_training)

print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in mean productivity scores.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in mean productivity scores.")
```

#### Decide and conclude

## Q2

 A researcher wants to investigate whether a new teaching method improves students' test scores. The researcher selects a random sample of 30 students and administers a test before and after implementing the teaching method. The scores (out of 100) are as follows:
 
Before: 75, 82, 68, 90, 78, 85, 72, 80, 88, 92, 84, 79, 76, 81, 83, 70, 87, 74, 77, 86, 73, 90, 79, 75, 83, 85, 81, 87, 89, 84

After: 80, 85, 72, 94, 82, 88, 75, 84, 90, 94, 88, 80, 78, 82, 85, 75, 92, 77, 80, 90, 76, 92, 80, 78, 88, 84, 83, 91, 88, 82

Using a significance level of 0.05, conduct a paired sample t-test to determine if there is a significant difference in test scores before and after the teaching method.

#### Solution

#### Define your hypothesis

H0: There is no significant difference in the mean test scores before and after implementing the teaching method.

H1: There is a significant difference in the mean test scores before and after implementing the teaching method.

#### Create lists for the samples

```python
before = [ 75, 82, 68, 90, 78, 85, 72, 80, 88, 92, 84, 79, 76, 81, 83, 70, 87, 74, 77, 86, 73, 90, 79, 75, 83, 85, 81, 87, 89, 84]
after = [80, 85, 72, 94, 82, 88, 75, 84, 90, 94, 88, 80, 78, 82, 85, 75, 92, 77, 80, 90, 76, 92, 80, 78, 88, 84, 83, 91, 88, 82]
```

#### Check the sample distrbuition using kde plot

```python
#Let us check their distrbuition
sns.kdeplot(before, color="blue", fill=True)
sns.kdeplot(after, color="red", fill=True)

plt.show()
```

#### Use scipy.stats to calcute paired sample t - test

```python
#calculate Paired sample t-test
from scipy.stats import ttest_rel

t_statistic, p_value = ttest_rel(before, after)

print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in test scores before and after the teaching method.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in test scores before and after the teaching method.")
```

## ANOVA (Analysis of Variance) 

A researcher wants to compare the effects of three different types of fertilizer on the growth of corn plants. He randomly assigns 15 plots of land to one of the three fertilizers: A, B, or C. After 12 weeks, he measures the height of the corn plants in each plot. The data are shown below:

Fertilizer  ----    	     Height (cm)

A ---	                120, 125, 130, 135, 140

B ---	                110, 115, 120, 125, 130

C --	                100, 105, 110, 115, 120


#### Define Hypothesis:


H0: There is no significant difference in the mean height of corn plants among the three types of fertilizers (A, B, and C).

H1: At least one pair of fertilizers has a significantly different mean height of corn plants.

#### Create lists for the samples

```python
fertilizer_A = [120, 125, 130, 135, 140]
fertilizer_B = [110, 115, 120, 125, 130]
fertilizer_C = [100, 105, 110, 115, 120]
```

#### Check the sample distrbuition using kde plot

```python
#Let us check their distrbuition
sns.kdeplot(fertilizer_A, color="green", fill=True)
sns.kdeplot(fertilizer_B, color="blue", fill=True)
sns.kdeplot(fertilizer_C, color="yellow", fill=True)
plt.show()
```

#### Use scipy.stats to calculate ANOVA F- test 

```python
#Calculate ANOVA 
from scipy.stats import f_oneway

f_statistic, p_value = f_oneway(fertilizer_A, fertilizer_B, fertilizer_C)

print(f"F-Statistic: {f_statistic}")
print(f"P-Value: {p_value}")

alpha = 0.05 

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in the mean height of corn plants among the fertilizers.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the mean height of corn plants among the fertilizers.")
```
