```python
#Load diamonds dataset from seaborn library (1 Mark)
kiki = sb.load_dataset('diamonds')
kiki
```

```python
# Show the first 10 rows from your data
kiki.head(10)
```

```python
#What the size of this dataset

print(f"Number of rows: {kiki.shape[0]}")
print(f"Number of columns: {kiki.shape[1]}")
```

#### We will work with depth column in this dataset

```python
#Draw the boxplot of depth column (1 Mark)

plt.figure(figsize=(8, 6))
plt.boxplot(kiki['depth'], vert=False)
plt.title('Boxplot of Depth')
plt.xlabel('Depth')
plt.show()
```

```python
#Calculate Quantile 1  (0.5 Mark)

Q1 = kiki['depth'].quantile(0.25)

print(f"The Q1 is: {Q1}")
```

```python
#Calculate Quantile 3 (0.5 Mark)
Q3 = kiki['depth'].quantile(0.75)

print(f"The Q3 is: {Q3}")
```

```python
#Calculate IQR (0.5 Mark)
IQR = Q3 - Q1

print(f"The IQR is: {IQR}")
```

```python
#Calculate the lower limit of the outlier (low outlier) (1 Mark)

k5 = Q1 - 1.5 * IQR

print(f"The lower limit of the outliers (low outlier) is: {k5}")
```

```python
#Calculate the upper limit of the outlier (high outlier) (0.5 Mark)

k6 = Q3 + 1.5 * IQR

print(f"The upper limit of the outliers (high outlier) is: {k6}")
```

```python
#Print IQR, low outlier, and high outlier

print(f"The IQR is: {IQR}")

print(f"The lower limit of the outliers (low outlier) is: {k5}")

print(f"The upper limit of the outliers (high outlier) is: {k6}")
```

```python
#Get the values of the dataset that are below the lower limit (1 Mark)

low_outliers = kiki[kiki['depth'] < k5]

print("Values below the lower limit (low outliers):")
print(low_outliers)


#print them
```

```python
#Get the values of the dataset that are above the upper limit (1 Mark)

high_outliers = kiki[kiki['depth'] > k6]

print("Values above the upper limit (high outliers):")
print(high_outliers)

#print them
```

```python
#Combine all the outlier values in one dataset (2 Mark)

all_outliers = pd.concat([low_outliers, high_outliers])

print(f"The shape of the outlier dataset is: {all_outliers.shape}")

#show the shape of the outlier
```

```python
#Drop the outliers rom the orginal dataset (2 Mark)

no_outliers = kiki.drop(all_outliers.index)

print(f"The shape of the dataset without outliers is: {no_outliers.shape}")


#Show the shape
```

```python
#Dra boxplot of the depth column in the cleaned dataset (1 Mark)

plt.figure(figsize=(8, 6))
plt.boxplot(no_outliers['depth'], vert=False)
plt.title('Boxplot of Depth (Cleaned Dataset)')
plt.xlabel('Depth')
plt.show()
```

```python
#Replace the outliers ith the median of depth column  (2 Mark)

median_depth = kiki['depth'].median()

kiki.loc[all_outliers.index, 'depth'] = median_depth
```

```python
#Dra boxplot of the depth column for this dataset (1 Mark)

plt.figure(figsize=(8, 6))
plt.boxplot(kiki['depth'], vert=False)
plt.title('Boxplot of Depth')
plt.xlabel('Depth')
plt.show()
```
