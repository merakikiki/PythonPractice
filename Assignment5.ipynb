### 1- What is the Probability that the Car is Sedan ?

### 6-  4a) Replace the null value of column "wheelbase" with the max value of the same column (1 mark)

```python
df = df.fillna({'carwidth': df['carwidth'].min() })

df.isnull().sum()
```

###  8- 4c) Use Linear Interpolation to change the null value of column "carheight"  (1 mark)

```python
df['carheight']= df['carheight'].interpolate()

df.isnull().sum()
```

### 9- 4d) Replace the null value of column "enginesize" with the mod of the same column (1 mark)

```python
enginesize_mode = df['enginesize'].mode()[0]

df['enginesize'].fillna(enginesize_mode, inplace=True)

df.isnull().sum()
```

### 10-  Given that the car height is higer than 50, find the probability that the car engin size is greater than 130? (1 mark)

```python
engine130_and_50 = len(df[(df['enginesize'] > 130) & (df['carheight'] > 50)])

cars_50 = len(df[df['carheight'] > 50])

condition_prob = engine130_and_50 / cars_50

condition_prob
```
